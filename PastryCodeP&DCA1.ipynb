```python
import numpy as np
```

```python
nodes=[]
```

```python
def genNodeId():
  # actually nodeId generated by using hash of ip address and it contains 32bits but due to limitations of resources we are making this with limits
  nodeId=np.random.randint(10000)
  flag=False
  while flag==False:
    if(len(nodes)==0):
      flag=True
    for i in range(len(nodes)):
      if(nodes[i][1]==nodeId):
        nodeId=nodeId+1
        flag=False
        break
      else:
        flag=True

  return nodeId
```

```python
def checkPrime(n):
  if n == 1 or n == 0:
      return 0
  for i in range(2, n//2):
      if n % i == 0:
          return 0
  return 1
def getPrime(n):
    if n % 2 == 0:
        n = n + 1
    while not checkPrime(n):
        n += 2
    return n
def hashFunction(key):
    capacity = getPrime(10)
    return key % capacity
```

```python
def addNode():
  nodeId=genNodeId()
  keyId=genNodeId()
  node=[]
  node.append(nodeId)
  node.append(keyId)
  nodes.append(node)
  return nodes
```

```python
def addData(key,data):
  for i in range(len(nodes)):
    if(nodes[i][1]==key):
      nodes[i].append(data)
```

```python
def showData():
  for i in range(len(nodes)):
    print(nodes[i])
```

```python
def remNode(id):
  for i in range(len(nodes)):
    if(id==nodes[i][1]):
      nodes.pop(i)
    else:
      print("Node not found")
```
